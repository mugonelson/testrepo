## Tasks

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
query = "SELECT * FROM SPACEXTABLE;"
cur.execute(query)

# Fetch and print the unique rows
unique_values = cur.fetchall()
for row in unique_values:
    print(row)
```

```python
cur.execute("SELECT * FROM SPACEXTABLE LIMIT 0")
column_names = [description[0] for description in cur.description]
print(column_names)
```

```python
query = "SELECT DISTINCT Launch_Site FROM SPACEXTABLE;"
cur.execute(query)

# Fetch and print the unique rows
unique_values = cur.fetchall()
for row in unique_values:
    print(row[0])
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
searchterm = "CCA"
queryparam=f"%{searchterm}%"


query = "SELECT * FROM SPACEXTABLE WHERE Launch_Site like ? LIMIT 5"
cur.execute(query, (queryparam,))
results = cur.fetchall()
for row in results:
    print(row)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
searchterm = "NASA (CRS)"
queryparam=f"%{searchterm}%"


query = "SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Customer like ?"
cur.execute(query, (queryparam,))
results = cur.fetchall()
for row in results:
    print(row[0])
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
searchterm = "F9 v1.1"
queryparam=f"%{searchterm}%"


query = "SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Booster_Version like ?"
cur.execute(query, (queryparam,))
results = cur.fetchall()
for row in results:
    print(row[0])
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
searchterm = "Success (ground pad)"
queryparam=f"%{searchterm}%"


query = "SELECT MIN(Date) FROM SPACEXTABLE WHERE Landing_Outcome like ?"
cur.execute(query, (queryparam,))
results = cur.fetchall()
for row in results:
    print(row)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
searchterm = "Success (drone ship)"
queryparam=f"%{searchterm}%"


query = "SELECT Booster_Version, PAYLOAD_MASS__KG_ FROM SPACEXTABLE WHERE Landing_Outcome like ? AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000"
cur.execute(query, (queryparam,))
results = cur.fetchall()
for row in results:
    print(row)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
searchterm = "Success"
queryparam=f"%{searchterm}%"


query = "SELECT * FROM SPACEXTABLE WHERE Mission_Outcome like ?"
cur.execute(query, (queryparam,))
results = cur.fetchall()
print(f"total number of successful mission outcomes are: {len(results)}")
for row in results:
    print(row)
```

```python
searchterm = "Success"
queryparam=f"%{searchterm}%"


query = "SELECT * FROM SPACEXTABLE WHERE Mission_Outcome like ?"
cur.execute(query, (queryparam,))
results = cur.fetchall()
print(f"total number of successful mission outcomes are: {len(results)}")
for row in results:
    print(row)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
query = "SELECT Booster_Version, MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE GROUP BY Booster_Version"
cur.execute(query)
results = cur.fetchall()
for row in results:
    print(row)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
searchterm = "Failure (drone ship)"
queryparam=f"%{searchterm}%"

searchterm2 = "2015"
queryparam2=f"%{searchterm2}%"

query = "SELECT substr(Date, 6,2) as Month, Booster_Version, Launch_Site, Landing_Outcome FROM SPACEXTABLE WHERE Landing_Outcome like ? and Date Like ?"
cur.execute(query, (queryparam, queryparam2,))
results = cur.fetchall()
for row in results:

    print(row)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
searchterm = '2010-06-04'
searchterm2 = '2017-03-20'


query = "SELECT Date, Booster_Version, Launch_Site, Payload, PAYLOAD_MASS__KG_, Orbit, Customer, Mission_Outcome, Landing_Outcome, COUNT(Landing_Outcome) FROM SPACEXTABLE WHERE Date BETWEEN ? AND ? GROUP BY Landing_Outcome ORDER BY Date DESC"
cur.execute(query, (searchterm, searchterm2,))
results = cur.fetchall()
for row in results:
    print(row)
```
